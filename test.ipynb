## XPATH FOR CATEGORIES

### All Categories

```python
import requests 
from lxml import html

r = requests.get('https://www.agryco.com')
doc = html.fromstring(r.content)
print(doc)
print(doc.xpath('//li[contains(@class, "nav-item   nav-parent-item")]//li[contains(@class, "nav-item ")]/a[@class="nav-link"]/@href'))
```

### For "Promos" Category

```python
import requests 
from lxml import html

r = requests.get('https://www.agryco.com')
doc = html.fromstring(r.content)
print(doc)
print(doc.xpath('//li[@id="menu-13"]/a/@href'))
print(doc.xpath('normalize-space(//li[@id="menu-13"]//span)'))
```
